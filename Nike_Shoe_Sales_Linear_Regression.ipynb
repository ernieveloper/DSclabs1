# Step 1



```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb

nike_sales = pd.read_csv('nike_shoes_sales.csv')
nike_sales.info()
```

```python
nike_sales.describe()
```

```python
nike_sales.head()
```

# Step 2

Looking for the correlation of the Data

```python
Corr_nike_Sales = nike_sales.select_dtypes(include=['int64','float64'])
Corr_nike_Sales.head()
matrix = Corr_nike_Sales.corr()
sb.heatmap(matrix, annot= True, cmap ='coolwarm')

```

```python
sb.pairplot(Corr_nike_Sales)
```

# Step 3

Seems that there is a correlation between listing_price and sales_price, cleanning the data and rerun the coorelation

```python
Corr_nike_Sales = nike_sales.select_dtypes(include=['int64','float64'])
Corr_nike_Sales2 = Corr_nike_Sales.loc[Corr_nike_Sales['listing_price']>0]
Corr_nike_Sales2.head()
matrix2 = Corr_nike_Sales2.corr()
sb.heatmap(matrix2, annot= True, cmap ='coolwarm')
```

```python
sb.pairplot(Corr_nike_Sales2)
```
